```python
import cv2
import numpy as np
import random
from matplotlib import pyplot as plt


image = cv2.imread('a.jpg')
#meanfilter
mean_filtered = cv2.blur(image, (5, 5))
cv2.imwrite('a_mean_filter.jpg', mean_filtered)
#bien
edges = cv2.Canny(image, 100, 200)
cv2.imwrite('a_edge.jpg', edges)
#kenhmauRGb
channels = list(cv2.split(image))  # Chuyn tuple thnh list
random.shuffle(channels)          # Sau  mi shuffle c
random_color_image = cv2.merge(channels)
cv2.imwrite('a_random_color.jpg', random_color_image)
#hsv
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
cv2.imwrite('a_hue.jpg', h)
cv2.imwrite('a_saturation.jpg', s)
cv2.imwrite('a_value.jpg', v)

plt.figure(figsize=(15, 10))

# nh gc
plt.subplot(3, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("nh gc (BGR)")
plt.axis('off')

# Mean filter
plt.subplot(3, 3, 2)
plt.imshow(cv2.cvtColor(mean_filtered, cv2.COLOR_BGR2RGB))
plt.title("Mean Filter")
plt.axis('off')

# Edge detection
plt.subplot(3, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis('off')

# Random color
plt.subplot(3, 3, 4)
plt.imshow(cv2.cvtColor(random_color_image, cv2.COLOR_BGR2RGB))
plt.title("Random BGR Color")
plt.axis('off')

# Hue channel
plt.subplot(3, 3, 5)
plt.imshow(h, cmap='gray')
plt.title("Hue Channel (HSV)")
plt.axis('off')

# Saturation channel
plt.subplot(3, 3, 6)
plt.imshow(s, cmap='gray')
plt.title("Saturation Channel (HSV)")
plt.axis('off')

# Value channel
plt.subplot(3, 3, 7)
plt.imshow(v, cmap='gray')
plt.title("Value Channel (HSV)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
