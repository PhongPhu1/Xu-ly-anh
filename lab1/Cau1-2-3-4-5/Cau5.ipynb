```python
import numpy as np
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt

# Đọc ảnh ở chế độ grayscale (float)
a = iio.imread('baby.jpeg', mode='F')
b = iio.imread('balloons_noisy.png', mode='F')
c = iio.imread('flower.jpeg', mode='F')

# Tạo kernel trung bình 5x5
k = np.ones((5, 5)) / 25

# Áp dụng bộ lọc trung bình
d = sn.convolve(a, k).astype(np.uint8)
e = sn.convolve(b, k).astype(np.uint8)
f = sn.convolve(c, k).astype(np.uint8)

# Lưu ảnh sau lọc
iio.imsave('baby_mean.jpeg', d)
iio.imsave('balloons_noisy_mean.png', e)
iio.imsave('flower_mean.jpeg', f)

# Hiển thị ảnh bằng matplotlib
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(d, cmap='gray')
axes[0].set_title('Baby Image (Mean Filtered)')
axes[0].axis('off')

axes[1].imshow(e, cmap='gray')
axes[1].set_title('Balloons Noisy Image (Mean Filtered)')
axes[1].axis('off')

axes[2].imshow(f, cmap='gray')
axes[2].set_title('Flower Image (Mean Filtered)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
