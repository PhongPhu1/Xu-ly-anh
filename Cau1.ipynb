```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt

img = iio.imread('bird.png')

red_img = np.zeros_like(img)
green_img = np.zeros_like(img)
blue_img = np.zeros_like(img)

red_img[:, :, 0] = img[:, :, 0]
green_img[:, :, 1] = img[:, :, 1]
blue_img[:, :, 2] = img[:, :, 2]

iio.imwrite('bird_red.png', red_img)
iio.imwrite('bird_green.png', green_img)
iio.imwrite('bird_blue.png', blue_img)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(red_img)
axs[1].imshow(green_img)
axs[2].imshow(blue_img)

plt.show()
```
